```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
import seaborn as sns 
sns.set(color_codes=True)
%matplotlib inline
```

```python
data = pd.read_csv('indian_food.csv' )
```

```python
data
```

# Cleaning Data 

```python
# here we are going to sort the values of the name of the item in the alphabetical order to get data look nice 
data.sort_values(by='name' , inplace = True )
```

```python
data.head(15)
```

```python
# Here i am going to add a new column to Dataframe so that we can make it the index of the dataframe 

# python code to print the list of number raning from 0 to 255 that is the number of our rows
index_column = [i for i in range(255)]

data['index'] = index_column # code to add the list as column in dataframe
data
```

```python
# here i am going to make the index column as my index of the dataframe
data.set_index('index', inplace=True)
```

```python
data.head(50)
```

## Big Conditional Analysis

```python
# here we are going to calculate the data of the fastest dish (less than half hour) to be prepare and it must be from maharashtra , it must be spicy , it must be in maincourse and it must be vegetarian

data[(data['cook_time'] < 60 ) & (data['state'] == 'Maharashtra') & (data['flavor_profile'] == 'spicy') & (data['course'] == 'main course') &(data['diet'] == 'vegetarian')]
```

# Visualizing the Dataframe

```python
# plotting the data in the seaborn type
# here we can see that only 1 food need more than 500 minutes that is Shrikhand
# here KDE is the line of the density analyzer kde = Kernel Density Estimation
sns.distplot(data['cook_time'])
plt.show()
```

```python
# this plot shows the best time need for the cooking of a food item
# here we can see that only 1 food need more than 500 minutes that is Shrikhand
# here KDE is flase means off
sns.displot(data['cook_time'], kde=False )
plt.show()
```

```python
# here we are going to create the bi variate distribution visualization
# this also say the comapring plot with diffrent aspect
# here i have compare the maharashtra state with other state in count of dishes which is calculated by the index
sns.jointplot(data['state'] == 'Maharashtra' , data['state'])
plt.show()
```

```python
sns.jointplot(data['state'] == 'Maharashtra' , data['state'] == 'Gujarat' , kind='kde')
plt.show()
```

```python
# here we are going to create the multivariate visualization
sns.pairplot(data[['cook_time' , 'prep_time']])
plt.show()
```

```python
# create the 2D plot 
sns.stripplot(data['flavor_profile'], data['state']  , jitter=True)
plt.show()
```

```python
# create the 2D plot 
# swarmplot
sns.swarmplot(data['flavor_profile'], data['state'])
plt.show()
```

```python
sns.boxplot(data['cook_time'], data['prep_time'] ,  hue=data['course'] )
plt.show()
```

```python
sns.barplot(data['cook_time'], data['prep_time'] ,  hue=data['course'] )
plt.show()
```

```python
sns.countplot(data['region'] , hue=data['cook_time'])
plt.show()
```

```python
sns.pointplot(data['state'] == 'Maharashtra' , data['course'] ,hue=data['cook_time'])
plt.show()
```

```python
# kind : point , bar, count , box , violin , strip , boxen
sns.factorplot(x='diet', y='prep_time' , hue='cook_time' , col='region' , data=data , kind='swarm')

plt.show()
```

```python
# plotting a linear regression plot
sns.lmplot(y='cook_time' , x='prep_time' , data=data , hue='region')
plt.show()
```

```python
plt.style.use('seaborn')

flavor_profile_set = data['flavor_profile']
cooking_time = data['cook_time']
plt.plot(cooking_time , flavor_profile_set , color='m' , linestyle='--'  ,label='cook' )

preparation_time = data['prep_time']
plt.plot(preparation_time ,flavor_profile_set, color='r' , linestyle='--'  ,label='prep' )


plt.xlabel('Time') # labeling the X axis 
plt.ylabel('State') # labeling the y axis 
plt.title('Indian Food') # Naming the chart 

plt.legend() # this will help to understand the about the lines on the graph

plt.grid(True) # add the grid to understand the graph


plt.tight_layout() # this will auto adjust the lines in the graph


plt.show()
```
